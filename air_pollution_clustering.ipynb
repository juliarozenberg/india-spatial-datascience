# 1. Introduction & Problem statement

When imagining air pollution, one will often picture smog in big Indian cities, crowned by a red sun. It is said that in 2019, 21 out of 30 of the most polluted cities in the world were located in the Indian subcontinent (1). Air pollution can have both severe short-term and long-term effects on people's lives and has been linked not only to lung diseases but heart diseases as well (2). 

There are many causes of air pollution, ranging from dust to burning of waste, transportation caused pollution and heavy industry's emissions. However, there also seems to be a spatial-geographic component to air pollution - a region's geography (i.e. mountain ranges) might exarcerbate air pollution, whilst a sea-breeze can help alleviate air pollution locally. Further, it is generally accepted that vegetation is able to filter out air pollution.


![Wikipedia causes of Indian Air pollution](https://upload.wikimedia.org/wikipedia/commons/2/2d/Sources_of_Air_Pollution.png)

Therefore, I hypothesise that districts in India with relatively low air pollution will have more vegetation coverage (since plants filter out air pollution) and less urban-centric human activities. I will investigate the relation between air pollution and 6 other variables, which are listed underneath. Further, by considering the spatial occurence of these variables, I will try to make a connection with the local geography (mountain ranges, seaside, etc.).



**Definitions of metrics:**
I will measure these effects at district level for all variables.
* Surface PM2.5 pollution (estimated annual ground-level fine particulate matter (PM2.5))
* Forest coverage using mean VCF data.

Metrics for human-activities:

#### i) Surface PM 2.5

````python
# since the loading of data is repetitive, and we just want to see the columns & rows for the different data, I have used this function to do so. 
#There were 640 districts in 2011, so the datasets should have 640 or less, but not more than 640 rows.

def shrug_load(file_name, delimiter=','):
    """
    Load shrug data, saves as pd_df, prints nrows, ncols and prints .head()
    """
    file_path = "data/"
    full_path = file_path + file_name
    data = pd.read_csv(full_path, delimiter=delimiter)

    print(f"The {file_name.split('.')[0]} Data")
    print(f"Rows: {data.shape[0]}")  # Number of rows
    print(f"Columns: {data.shape[1]}")  # Number of columns
    
    return data
````

## 4.2 Clustering without prior PCA

Using kmeans method, on standardised data! First I determine the right K (amount of clusters) using the elbow method, then use kmeans to create the clusters and analyse them.

````python
features_c = ['pm25_mean', 'vcf_mean', 'tot_p', 'tot_p_manufacturing', 'tot_len_roads', 'tot_el_dom'] # we do not use the binary value b_waste for clustering here!, so list for new offset

X = standard_gdf[features_c] # use standardized variables! 
X # view the offset dataset, excluding geometry data
````

**Perform elbow method for finding right k**

````python
wcss = [] 

k_scale = 11

for i in range(1, k_scale): 
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(X) 
    wcss.append(kmeans.inertia_)

x = np.arange(1, k_scale)

plt.plot(x, wcss, 'bo-')
plt.title('Plot for elbow method on k-means')
plt.xlabel('K')
plt.xticks(x)
plt.ylabel('Within-Cluster Sum of Squares')
plt.plot(4, wcss[3], 'ro', label='Optimal K value for k-means clustering')
plt.legend()

;
````

- Best K for K means clustering, based on elbow method, is here K = 4
- This is subjective, but since I want to keep the interpretability high, I decide on a lower K (I have also considered a k of 5).

## 4.3 PCA

The clustering in 4.2 was performed using 5 different features, by using PCA I can reduce the dimensionality of the data. PCA is performed on the standardized data. I used this website as a reference: # https://www.kindsonthegenius.com/principal-components-analysispca-in-python-step-by-step/ 

````python
from sklearn.decomposition import PCA
````

````python
features_pca = ['vcf_mean', 'tot_p', 'tot_p_manufacturing', 'tot_len_roads', 'tot_el_dom', 'b_waste' ]

X = standard_gdf[features_pca] # use the scaled features dataframe
y = standard_gdf['pm25_mean']
X.describe()
````

Multicollinearity in set of predictors?
- tot_p & tot_p_manufacturing
- tot_p & tot_len_roads
- tot_el_dom & tot_len_roads
- tot_p & tot_el_dom
<br>

Now, let's start the PCA, First by choosing a max of 5 components (more would not make sense since I currently have 5 features)

````python
pca = PCA(n_components=5) # 
columns = ['PC1', 'PC2', 'PC3', 'PC4', 'PC5']

principalComponents = pca.fit_transform(X) # run algorithm

principalDataframe = pd.DataFrame(data = principalComponents, columns=columns ) # create new df containing PCA Data

pca_df = principalDataframe.set_index(X.index) # set index to be the same as for other dataset to make merging later easier
pca_df # display df
````

````python
targetDataframe = standard_gdf['pm25_mean'] # add PM25 (dependent variable) to dataframe
newDataframe = pd.concat([pca_df, targetDataframe], axis = 1)
newDataframe
````

````python
percent_variance = np.round(pca.explained_variance_ratio_* 100, decimals =2) # for plotting the histogram, defines bounds
num_pcs = len(percent_variance)
````

**Decide on amount of PCs**

<br>

I will do this using the elbow method

````python
x = range(1, num_pcs + 1)
plt.bar(x, height=percent_variance, tick_label=columns)

plt.plot(x, percent_variance, color='yellow', marker='o', linestyle='--', label='Percentage of Variance')
plt.plot(x[2], percent_variance[2], color='red', marker='o', label='Chosen amount of PCs with elbow method')
plt.ylabel('Percentate of Variance Explained')
plt.xlabel('Principal Component')
plt.title('PCA Scree Plot')
plt.legend()

plt.show()
````

- Here: decide on 3 principal Components, since together they may explain 80% of variance and is easier for interpretability!

````python
pc3_gdf = pd.concat([standard_gdf, newDataframe[['PC1', 'PC2', 'PC3']]], axis=1) # save 3 principal components in new dataframe
pc3_gdf
````

Using PCA Loadings and Correlation

````python
loadings = pd.DataFrame(
    pca.components_, 
    columns=features_pca, 
    index=['PC1', 'PC2', 'PC3', 'PC4', 'PC5']
)
print("PCA Loadings:")
loadings
````

Now visualize them, makes it easier to analyse

````python
def plot_pca_loadings(loadings, pc_name):
    # Extract and sort loadings for the specified PC
    loadings = loadings.T
    pc_loadings = loadings[pc_name].sort_values(ascending=False)
    # Plot as a bar chart
    plt.figure(figsize=(6, 3))
    plt.bar(pc_loadings.index, pc_loadings.values)
    plt.xlabel('Variables')
    plt.ylabel('Loadings')
    plt.title(f'Top Variables Contributing to {pc_name}')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
````
