## 2.3 Analysis & Interpretation

I would like to explore my hypothesis by looking at some chloropleths of the population, forest coverage and air pollution data. Are the districts with a lot of forest coverage, the same as the least polluted areas? Do these districts have a great amount of inhabitants? First, for flexibility, I write 2 functions to i) plot major rivers and urban centres and ii) the data from the previously created gdf. Since it concerns continuous data, I have to make use of some method of categorizing the data to colour codes.

````python
def plot_rivers_cities(ax):
    """ 
    Load and plot rivers and cities of India on an ax (input). Uses epsg=4326 and data is from https://www.data.gov.in/catalog/boundaries-water-resources-projects?page=2'
    It is assumed they are stored under the path ./data/shapefiles/... relative to the location of this JupyterNB File. 
    """ 
    path = './data/shapefiles/urban' 
    urban_gdf = gpd.read_file(path)
    urban_gdf.to_crs(epsg=4326, inplace=True)

    path = './data/shapefiles/rivers'
    rivers_gdf = gpd.read_file(path)
    rivers_gdf.to_crs(epsg=4326, inplace=True)
    
    rivers_gdf.plot(linewidth=0.4, color='blue' ,ax=ax)
    urban_gdf.plot(color='red', markersize=0.1, ax=ax, alpha=0.7)
    return
````

````python
def plot_shrug_data(gdf, column, legend_title, scheme='Quantiles', k=7, cmap=plt.cm.Blues):
    """ 
    Plots the gdf from the shrug database. In the funtion, the column to be plotted, scheme, k and colormap may be altered. other setting are set in the function
    """
    f, ax = plt.subplots(1) # always plots 1 figure
    gdf.plot(
        column=column,
        scheme=scheme,
        k=k,
        cmap=cmap,
        ax=ax,
        alpha=1,
        edgecolor="w",
        linewidth=0.1,
        legend=True,
        legend_kwds={
        'loc': 'upper left',          # Position legend outside the plot
        'bbox_to_anchor': (1, 1),    # Adjust position of the legend box
        'title': legend_title,       # Legend title
        'fmt': "{:.0f}"              # Format labels to remove decimals
        }
    )
    plot_rivers_cities(ax)    
    return
````
